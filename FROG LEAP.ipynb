```python
def display_current_state(positions):
    print("[ 0 ,  1 ,  2 ,  3 ,  4 ,  5 ,  6 ]")
    print(positions)


def frog_leap_game():
    frog_positions = ['G', 'G', 'G', '-', 'B', 'B', 'B']
    display_current_state(frog_positions)

    while True:
        user_input = input("Press 'q' to quit or enter the position of the frog: ")

        if user_input.lower() == 'q':
            print("You decided to end the game :( .")
            break

        selected_position = int(user_input)

        if selected_position < 0 or selected_position > 6:
            print("Invalid Move. Please enter a position between 0 and 6.")
            continue

        empty_position = frog_positions.index('-')

        if frog_positions[selected_position] == 'G':
            if selected_position + 1 <= 6 and frog_positions[selected_position + 1] == '-':
                empty_position = selected_position + 1
            elif (
                selected_position + 2 <= 6
                and frog_positions[selected_position + 2] == '-'
                and frog_positions[selected_position + 1] == 'B'
            ):
                empty_position = selected_position + 2
            else:
                print("Invalid Move. Try again.")
                continue
        elif frog_positions[selected_position] == 'B':
            if selected_position - 1 >= 0 and frog_positions[selected_position - 1] == '-':
                empty_position = selected_position - 1
            elif (
                selected_position - 2 >= 0
                and frog_positions[selected_position - 2] == '-'
                and frog_positions[selected_position - 1] == 'G'
            ):
                empty_position = selected_position - 2
            else:
                print("Invalid Move. Try again o_0 .")
                continue

        frog_positions[selected_position], frog_positions[empty_position] = (
            frog_positions[empty_position],
            frog_positions[selected_position],
        )
        display_current_state(frog_positions)

        if frog_positions == ['B', 'B', 'B', '-', 'G', 'G', 'G']:
            print("Congratulations! You Win! :D ")
            break

frog_leap_game()
```
